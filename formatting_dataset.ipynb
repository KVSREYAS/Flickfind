```python
import datasets
import numpy as np
from transformers import BertTokenizerFast
from transformers import DataCollatorForTokenClassification
from transformers import AutoModelForTokenClassification
```

```python
from datasets import load_dataset, Dataset

def load_conll_data(file_path):
    sentences = []
    labels = []

    with open(file_path, 'r') as file:
        sentence = []
        label = []
        for line in file:
            if line.strip() == "":
                if sentence:
                    sentences.append(sentence)
                    labels.append(label)
                    sentence = []
                    label = []
            else:
                parts = line.strip().split('\t')
                if len(parts) == 2:
                    sentence.append(parts[1])
                    label.append(parts[0])
        if sentence:
            sentences.append(sentence)
            labels.append(label)

    return sentences, labels

def convert_to_hf_format(sentences, labels):
    data = []
    for sent, lab in zip(sentences, labels):
        data.append({"tokens": sent, "ner_tags": lab})
    return data

# Load your data
file_path = '/content/mitmoviecomplex.txt'
sentences, labels = load_conll_data(file_path)

# Convert to Hugging Face dataset format
data = convert_to_hf_format(sentences, labels)
dataset = Dataset.from_dict({"tokens": [item["tokens"] for item in data], "ner_tags": [item["ner_tags"] for item in data]})

# Print an example
print(dataset[1])

# If you want to save the dataset to disk
dataset.save_to_disk("ner_dataset")
```

```python
entity_map = {
    'O': 0,
    'B-Actor': 1,
    'I-Actor': 2,
    'B-Character_Name': 3,
    'I-Character_Name': 4,
    'B-Director': 5,
    'I-Director': 6,
    'B-Genre': 7,
    'I-Genre': 8,
    'B-Plot': 9,
    'I-Plot': 10,
    'B-Year': 11,
    'I-Year': 12,
    'B-Soundtrack': 13,
    'I-Soundtrack': 14,
    'B-Opinion': 15,
    'I-Opinion': 16,
    'B-Award': 17,
    'I-Award': 18,
    'B-Origin': 19,
    'I-Origin': 20,
    'B-Quote': 21,
    'I-Quote': 22,
    'B-Relationship': 23,
    'I-Relationship': 24
}
# Function to encode the entities in a dataset
def encode_entities(dataset):
    encoded_dataset = []
    for data in dataset:
      labels=data["ner_tags"]
      tokens=data["tokens"]
      encoded_labels = [entity_map[label] for label in labels]
      encoded_dataset.append((tokens, encoded_labels))
    return encoded_dataset
```

```python
def tokenize_and_align_labels(examples, label_all_tokens=True):
    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)
    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        # word_ids() => Return a list mapping the tokens
        # to their actual word in the initial sentence.
        # It Returns a list indicating the word corresponding to each token.
        previous_word_idx = None
        label_ids = []
        # Special tokens like `` and `<\s>` are originally mapped to None
        # We need to set the label to -100 so they are automatically ignored in the loss function.
        for word_idx in word_ids:
            if word_idx is None:
                # set –100 as the label for these special tokens
                label_ids.append(-100)
            # For the other tokens in a word, we set the label to either the current label or -100, depending on
            # the label_all_tokens flag.
            elif word_idx != previous_word_idx:
                # if current word_idx is != prev then its the most regular case
                # and add the corresponding token
                label_ids.append(label[word_idx])
            else:
                # to take care of sub-words which have the same word_idx
                # set -100 as well for them, but only if label_all_tokens == False
                label_ids.append(label[word_idx] if label_all_tokens else -100)
                # mask the subword representations after the first subword

            previous_word_idx = word_idx
        labels.append(label_ids)
    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```
